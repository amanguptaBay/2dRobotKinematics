```python
import numpy as np
from roboticstoolbox.robot.ERobot import ERobot
from math import pi

class RobotXML(ERobot):

    def __init__(self):

        links, name, urdf_string, urdf_filepath = self.URDF_read(
            "/Users/aman/Library/Mobile Documents/com~apple~CloudDocs/Work/UCSD/Research - Yip/2dRobotKinematics/2dJointArm.urdf"
        )

        super().__init__(
            links,
            name=name,
            urdf_string=urdf_string,
            urdf_filepath=urdf_filepath,
        )

        self.manufacturer = "File Import"

        # zero angles, upper arm pointing up, lower arm straight ahead
        self.addconfiguration("qz", np.array([0, 0, 0, 0]))

        # reference pose robot pointing upwards
        self.addconfiguration("up", np.array([0.0000, 0.0000, 1.5707, 0.0000]))

robot = RobotXML()
print(robot)
```
